```python
import pandas as pd
import os
```

```python
# ë°ì´í„°ì…‹ ê²½ë¡œ
input_file_path = "Dataset/3states/dataset_pa_with_embeddings.jsonl"
output_file_path = "Dataset/3states/dataset_pa_with_embeddings_vector.jsonl"

# ì²­í¬ì‚¬ì´ì¦ˆ
chunk_size = 5000

#ì„ë² ë”© ë²¡í„° ë“¤ì–´ìˆëŠ” ì»¬ëŸ¼
embedding_column = "embedding"

# ë“œëí•  ì»¬ëŸ¼ëª…
review_text = "text"
review_useful = "review_useful"
review_date = "date"

# ì˜ˆìƒ ì„ë² ë”© ì°¨ì›, ë‹¤ë¥´ë©´ ì•Œë¦¼
expected_dimension = 3072
```

```python
# ì´ì „ì— ì‹¤í–‰í•´ì„œ íŒŒì¼ì´ ë‚¨ì•„ìˆë‹¤ë©´ ì‚­ì œ
if os.path.exists(output_file_path):
    os.remove(output_file_path)
    print(f"ê¸°ì¡´ '{output_file_path}' íŒŒì¼ì„ ì‚­ì œí–ˆìŠµë‹ˆë‹¤.")

try:
    # ì„ë² ë”© ë°ì´í„°ì…‹ ì²­í¬ë‹¨ìœ„ë¡œ ì½ìŒ
    json_reader = pd.read_json(input_file_path, lines=True, chunksize=chunk_size)

    print(f"'{input_file_path}' íŒŒì¼ ì²˜ë¦¬ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤...")
    total_lines = 0

    # ê° ì²­í¬ë§ˆë‹¤
    for i, chunk in enumerate(json_reader):

        # embedding ì»¬ëŸ¼ ì°¨ì› í™•ì¸
        if embedding_column in chunk.columns:
            # ê° ì„ë² ë”© ë²¡í„°ì˜ ê¸¸ì´ë¥¼ ê³„ì‚°
            dimensions = chunk[embedding_column].apply(len)
            # ì˜ˆìƒ ì°¨ì›ê³¼ ë‹¤ë¥¸ í–‰ì´ ìˆëŠ”ì§€ í™•ì¸
            mismatched_rows = chunk[dimensions != expected_dimension]
            # ë¬¸ì œê°€ ìˆëŠ” í–‰ì´ í•˜ë‚˜ë¼ë„ ìˆë‹¤ë©´ ê²½ê³  ë©”ì‹œì§€ ì¶œë ¥
            if not mismatched_rows.empty:
                print(
                    f"{i+1}ë²ˆì§¸ ì²­í¬ì—ì„œ ì°¨ì›ì´ {expected_dimension}ì´(ê°€) ì•„ë‹Œ ë°ì´í„°ê°€ ë°œê²¬ë˜ì—ˆìŠµë‹ˆë‹¤."
                )
                print(
                    f"   -> ë¬¸ì œê°€ ìˆëŠ” í–‰ì˜ ì¸ë±ìŠ¤: {mismatched_rows.index.tolist()}"
                )
        else:
            print(f"{i+1}ë²ˆì§¸ ì²­í¬ì— '{embedding_column}' ì»¬ëŸ¼ì´ ì—†ìŠµë‹ˆë‹¤.")

        # í•„ìš”ì—†ëŠ” ì»¬ëŸ¼ ë“œë
        chunk_modified = chunk.drop(
            columns=[review_text, review_useful, review_date], errors="ignore"
        )

        # ìˆ˜ì •ëœ ì²­í¬ append
        chunk_modified.to_json(
            output_file_path,
            orient="records",
            lines=True,
            mode="a",
        )

        total_lines += len(chunk)
        print(f"{i+1}ë²ˆì§¸ ì²­í¬ ì²˜ë¦¬ ì™„ë£Œ. (ëˆ„ì  {total_lines} ë¼ì¸)")
    print(f"ìµœì¢… ê²°ê³¼ê°€ '{output_file_path}' íŒŒì¼ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

except FileNotFoundError:
    print(f"ì…ë ¥ íŒŒì¼ '{input_file_path}'ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
except Exception as e:
    print(f"ì˜¤ë¥˜ ë°œìƒ : {e}")
```
